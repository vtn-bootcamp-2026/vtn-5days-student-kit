```python
# Tự động hóa: Setup anonymizer-skill package và import module
import sys
import os
import shutil

# 1. Xác định thư mục templates nguồn
notebook_dir = os.getcwd()
possible_paths = [
    os.path.abspath(os.path.join(notebook_dir, "..")),
    os.path.abspath("."),
    os.path.abspath(".."),
    os.path.abspath("../../templates"),
]
template_dir = None
for p in possible_paths:
    if os.path.exists(os.path.join(p, "skills/anonymizer-skill/scripts/anonymizer.py")):
        template_dir = os.path.join(p, "skills/anonymizer-skill")
        break

if not template_dir:
    backup_dir = "/Users/shimazu/Documents/9. active/course_viettel_network/vtn-ai-builders-bootcamp-2026/03-practice/day-03/session-05-mini-tool-vibe-coding/templates"
    if os.path.exists(backup_dir):
        template_dir = os.path.join(backup_dir, "skills/anonymizer-skill")

# 2. Tự động hóa: Tạo các thư mục skill package trong outputs/vtn-session05
project_skill_dir = os.path.abspath(os.path.join(notebook_dir, "..", "..", "outputs", "vtn-session05", "anonymizer-skill"))
scripts_dst = os.path.join(project_skill_dir, "scripts")
kb_dst = os.path.join(project_skill_dir, "kb")
schemas_dst = os.path.join(project_skill_dir, "schemas")

os.makedirs(scripts_dst, exist_ok=True)
os.makedirs(kb_dst, exist_ok=True)
os.makedirs(schemas_dst, exist_ok=True)
print(f"📂 Đã chuẩn bị thư mục skill package tại: {project_skill_dir}")

# Copy files (ghi đè để luôn khôi phục/chuẩn bị trạng thái chuẩn hoạt động được)
if template_dir:
    # Copy scripts
    for f in ["anonymizer.py", "validator.py"]:
        src = os.path.join(template_dir, "scripts", f)
        dst = os.path.join(scripts_dst, f)
        if os.path.exists(src):
            shutil.copy(src, dst)
            print(f"✅ Đã chuẩn bị/khôi phục script: {f}")
    # Copy KB
    for f in ["pii-categories.md", "regex-patterns.md", "safe-terms.md"]:
        src = os.path.join(template_dir, "kb", f)
        dst = os.path.join(kb_dst, f)
        if os.path.exists(src):
            shutil.copy(src, dst)
            print(f"✅ Đã chuẩn bị/khôi phục KB: {f}")
    # Copy Schema
    src_schema = os.path.join(template_dir, "schemas", "anonymized-output.schema.json")
    dst_schema = os.path.join(schemas_dst, "anonymized-output.schema.json")
    if os.path.exists(src_schema):
        shutil.copy(src_schema, dst_schema)
        print("✅ Đã chuẩn bị/khôi phục schema json")
    # Copy root configs
    for f in ["SKILL.md", "skill.json"]:
        src = os.path.join(template_dir, f)
        dst = os.path.join(project_skill_dir, f)
        if os.path.exists(src):
            shutil.copy(src, dst)
            print(f"✅ Đã chuẩn bị/khôi phục config: {f}")
else:
    print("❌ Không tìm thấy thư mục templates nguồn.")

# 3. Import anonymizer module
if scripts_dst not in sys.path:
    sys.path.insert(0, scripts_dst)
try:
    from anonymizer import normalize_unicode, regex_redact
    print("\n✅ Đã import anonymizer thành công!")
    print(f"  - normalize_unicode: {normalize_unicode}")
    print(f"  - regex_redact: {regex_redact}")
except ImportError as e:
    print(f"\n❌ Lỗi import anonymizer: {e}")
```
